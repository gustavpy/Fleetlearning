```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os


```

```python
current_directory = os.getcwd()

# Move one level up from the current directory
parent_directory = os.path.abspath(os.path.join(current_directory, os.pardir))

# Define the path to the CSV file located one level above the current directory
csv_file_path = os.path.join(parent_directory, 'metadata.csv')

df = pd.read_csv(csv_file_path)
```

```python
df = df.drop(["frame_id", "time"], axis=1)
df.head(5)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))


# Country_code
df.value_counts("country_code").plot(kind="bar", ax=axes[0, 0])
axes[0, 0].set_title("Country Code")

# Collection_car
df.value_counts("collection_car").plot(kind="pie", ax=axes[0, 1])
axes[0, 1].set_title("Collection Car")

# Scraped_weather
df.value_counts("scraped_weather").plot(kind="pie", ax=axes[1, 0])
axes[1, 0].set_title("Scraped Weather")

# Road_condition
df.value_counts("road_condition").plot(kind="pie", ax=axes[1, 1])
axes[1, 1].set_title("Road Condition")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python

colors, _ = pd.factorize(df['collection_car'])
plt.scatter(df["longitude"], df["latitude"], c=colors, alpha=0.7)


plt.legend(['g', 'b', 'o', 'd', 'i'], loc='upper left')
plt.show()

```
